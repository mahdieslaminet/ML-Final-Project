```python
# ============================
# Iran Futsal Dataset Analysis
# ============================

# 1️⃣ Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set(style="whitegrid")

# 2️⃣ Load dataset
df = pd.read_csv('https://raw.githubusercontent.com/Zahra-Alikhani2004/ML_Final-project-/main/Futsal-IRAN-Dataset/futsal_iran_players.csv')

# 3️⃣ Display first 5 rows
print("First 5 rows of dataset:")
display(df.head())

# 4️⃣ Numeric summary
print("Numeric summary of the dataset:")
display(df.describe())

# 5️⃣ Map position names from Persian to English
position_map = {
    'مهاجم': 'Forward',
    'Pivot': 'Pivot',
    'فلنک راست': 'Right Winger',
    'فلنک چپ': 'Left Winger',
    'وینگر': 'Winger',
    'فوروارد': 'Forward',
    'دفاع': 'Defender',
    'مدافع چپ': 'Left Defender',
    'مدافع راست': 'Right Defender',
    'هافبک': 'Midfielder',
    'دروازه‌بان': 'Goalkeeper'
}
df['Position'] = df['پست'].map(position_map).fillna(df['پست'])

# 6️⃣ Count of Players by Position
print("Count of players by Position:")
position_counts = df['Position'].value_counts()
display(position_counts)

plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Position', palette="viridis")
plt.title("Number of Players by Position")
plt.xlabel("Position")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# 7️⃣ Height distribution
plt.figure(figsize=(8,5))
sns.histplot(df['قد (متر)'], bins=10, kde=True, color='skyblue')
plt.title("Height Distribution (m)")
plt.xlabel("Height (m)")
plt.ylabel("Count")
plt.show()

# 8️⃣ Weight distribution
plt.figure(figsize=(8,5))
sns.histplot(df['وزن (کیلوگرم)'], bins=10, kde=True, color='salmon')
plt.title("Weight Distribution (kg)")
plt.xlabel("Weight (kg)")
plt.ylabel("Count")
plt.show()

# 9️⃣ Starting age distribution
plt.figure(figsize=(8,5))
sns.histplot(df['سن شروع حرفه‌ای'], bins=10, kde=True, color='lightgreen')
plt.title("Starting Age Distribution")
plt.xlabel("Starting Age")
plt.ylabel("Count")
plt.show()

# 🔟 Speed, Agility, Endurance distributions
fig, axes = plt.subplots(1,3, figsize=(18,5))
sns.histplot(df['سرعت'], bins=10, kde=True, color='gold', ax=axes[0])
axes[0].set_title("Speed Distribution")
sns.histplot(df['چابکی'], bins=10, kde=True, color='orange', ax=axes[1])
axes[1].set_title("Agility Distribution")
sns.histplot(df['استقامت'], bins=10, kde=True, color='lightcoral', ax=axes[2])
axes[2].set_title("Endurance Distribution")
plt.show()

# 1️⃣1️⃣ Skill distributions: Pass, Dribble, Shoot
fig, axes = plt.subplots(1,3, figsize=(18,5))
sns.histplot(df['مهارت پاس'], bins=10, kde=True, color='lightblue', ax=axes[0])
axes[0].set_title("Passing Skill Distribution")
sns.histplot(df['مهارت دریبل'], bins=10, kde=True, color='mediumpurple', ax=axes[1])
axes[1].set_title("Dribbling Skill Distribution")
sns.histplot(df['مهارت شوت'], bins=10, kde=True, color='lightgreen', ax=axes[2])
axes[2].set_title("Shooting Skill Distribution")
plt.show()
```
